# Perceptron

Perceptron is an algorithm that separates only the linear data. In the 2D plane, the perceptron acts as a line to separate the data; in the 3D region, the perceptron acts as a plane to split the data into two areas.

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generate sample data
random.seed(42)
data = {
    "CGPA": [round(random.uniform(7.0, 10.0), 2) for _ in range(400)],
    "IQ": [random.randint(90, 150) for _ in range(400)],
}

data["Placed"] = [1 if cgpa > 6 and iq > 110 else 0 for cgpa, iq in zip(data["CGPA"], data["IQ"])]
```

```python
dataset = pd.DataFrame(data)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
import seaborn as sns

sns.scatterplot(data=dataset, x="CGPA", y="IQ", hue="Placed")
```

```python
X = dataset.iloc[:,:2]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.linear_model import Perceptron
perceptron_model = Perceptron()
```

```python
perceptron_model.fit(X,y)
```

```python
perceptron_model.coef_
```

```python
perceptron_model.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=perceptron_model, legend=2)
plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.title("Perceptron Decision Boundary")
```
